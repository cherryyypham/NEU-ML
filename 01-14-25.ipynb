## l_p norm calculation

```python
# without the predefined numpy function
def l_p_norm(v, p):
    # the "**" is not as efficient as np.pow
    x_norm = np.sum(np.abs(v) ** p) ** (1 / p)

    # numpy is really efficient in C code --> whatever available numpy fucntion is better than vanilla python functions
    x_norm = np.sum(np.pow(np.pow(np.abs(v), p), (1 / p)))
    return x_norm

l_p_norm(x, 2)
```
