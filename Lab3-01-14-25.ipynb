```python
import numpy as np

M = np.random.randn(4,4)
A = M.T @ M # Gram matrix - symmetric and positive definite

print(f"A shape: {A.shape}")
print(f"A is symmetric: {np.allclose(A, A.T)}")
```

```python
eigenvals, eigenvectors = np.linalg.eig(A)
eigenvals
```

```python
eigenvectors
```

By definition, an eigenvector v satisfies Av = $\lambda$ v - the matrix only scales it

```python
v0 == eigenvectors[:,0]
Av0 = A @ v0
lambda0_v0 = eigenvals[0] * v0
```

```python
assert np.allclose(Av0, lambda0_v0)
```
