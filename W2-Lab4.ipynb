```python
import numpy as np

x = np.array([1, 2, 3])

def f(x):
    return x @ x  # ||x||^2

def grad_f(x):
    return 2 * x  # Gradient of ||x||^2 is 2x

print(f"x = {x}")
print(f"f(x) = {f(x)}")
print(f"grad_f(x) = {grad_f(x)}")
```

```python
def numerical_gradient(f, x, epsilon=1e-7):
    """Approximate gradient using central differences."""
    grad = np.zeros_like(x)
    for i in range(len(x)):
        x_plus = x.copy()
        x_minus = x.copy()
        x_plus[i] += epsilon
        x_minus[i] -= epsilon
        grad[i] = (f(x_plus) - f(x_minus)) / (2 * epsilon)
    return grad
```

$$ x_{new} = x_{old} - \eta \cdot  \nabla f(x_{old})$$

```python
learning_rate = 0.1

for i in range(6):
    loss = f(x)
    grad = grad_f(x)
    x = x - learning_rate * grad
    print(f"Iteration {i}: loss = {loss}, x = {x}")
```
